# EDA  = Exploratory Data Analysis

```python
gen = sns.countplot(x="Gender",data = ds)
sns.set(rc={'figure.figsize':(5,5)},style='dark')
for bar in gen.containers:
    gen.bar_label(bar)
```

```python
sales_gen = ds.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(5,4)},style='whitegrid')
sns.barplot(x='Gender',y='Amount',data = sales_gen)
```

* From above graph we can see that most of buyer are Female and also purchasing power of female is more male. 

### Age

```python
age_gen = sns.countplot(x='Age Group',hue='Gender',data = ds)
sns.set(rc={'figure.figsize':(15,5)},style='whitegrid')
for bar in age_gen.containers:
    age_gen.bar_label(bar)
```

```python
age_amo = ds.groupby(['Age Group'],as_index = False)['Amount'].sum().sort_values(by = 'Amount',ascending = False)
sns.set(rc={'figure.figsize':(10,5)},style='whitegrid')
sns.barplot(x='Age Group',y='Amount',data=age_amo)
```

*From above graphs we can see that most of the buyers are of age group between 26-35 yrs female*

### State

```python
state_ord = ds.groupby(['State'],as_index=False)['Orders'].sum().sort_values('Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(16,5)},style ='whitegrid')
sns.barplot(x='State',y='Orders',data=state_ord)
```

```python
state_amo = ds.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(16,5)},style ='whitegrid')
sns.barplot(x='State',y='Amount',data=state_amo)
```

*From above graphs we can see that most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectively*

### Marital_Status

```python
mar_st = sns.countplot(x='Marital_Status',data=ds)
sns.set(rc={'figure.figsize':(3,2)},style='whitegrid')

for bar in mar_st.containers:
    mar_st.bar_label(bar)
```

```python
mar_gen = ds.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(5,3)},style='whitegrid')
sns.barplot (x='Marital_Status',y='Amount',hue="Gender",data=mar_gen)
```

*From above graphs we can see that most of the buyers are unmarried (women) and they have high purchasing power*

### Occupation

```python
ocp = sns.countplot(x='Occupation',data = ds)
sns.set(rc={'figure.figsize':(15,5)},style = 'whitegrid')
plt.xticks(rotation = 60)
for bar in ocp.containers:
    ocp.bar_label(bar)
```

```python
ocp_amo = ds.groupby('Occupation',as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(20,5)},style='whitegrid')
sns.barplot(x='Occupation',y='Amount',data=ocp_amo)

```

*From above graphs we can see that most of the buyers are working in IT, Healthcare and Aviation sector*

### Product Category

```python
pr_cat = sns.countplot(x='Product_Category',data=ds)
sns.set(rc={'figure.figsize':(24,5)},style='whitegrid')

for bar in pr_cat.containers:
    pr_cat.bar_label(bar)
```

```python
prodcat_sale = ds.groupby('Product_Category',as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={"figure.figsize":(20,5)},style='whitegrid')
sns.barplot(x='Product_Category',y='Amount',data=prodcat_sale)
```

*From above graphs we can see that most of the sold products are from Food, Clothing and Electronics category*

```python
product_sale = ds.groupby('Product_ID',as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(15)
sns.set(rc={'figure.figsize':(20,5)},style='whitegrid')
sns.barplot(x='Product_ID',y='Amount',data=product_sale)
```
